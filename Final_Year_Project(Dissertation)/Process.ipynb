```python
#Function for filtering each signal
def ECG_filtering(raw_signal, sample_rate):
    ecg_signal=nk.ecg_clean(raw_signal, sample_rate)
    return ecg_signal

def EMG_filtering(low_boundary,high_boundary,s_rate,signal_array):
    emg_meaned=signal_array-np.mean(signal_array)
    high_pass=low_boundary/(s_rate/2)
    low_pass=high_boundary/(s_rate/2)
    b, a = scipy.signal.butter(4, [high_pass,low_pass], btype='bandpass')
    emg_signal = scipy.signal.filtfilt(b, a, emg_meaned)
    return emg_signal
    #Diong, J. 2016. Python: Analysing EMG signals â€“ Part 3. Available at: https://scientificallysound.org/2016/08/18/python-analysing-emg-signals-part-3/ [Accessed:21 April 2025]

def EDA_filtering(raw_signal, sample_rate ):
    processed=eda.eda(signal=raw_signal, sampling_rate=sample_rate, show=False)  
    EDA_filtered=processed["filtered"]
    EDA_first_down=scipy.signal.decimate(EDA_filtered,10)   #reduce sample rate from 700 to 70
    EDA_second_down=scipy.signal.decimate(EDA_first_down,5) #reduce sample rate from 70 to 14hz
    
    EDA_min=np.min(EDA_second_down)
    EDA_max=np.max(EDA_second_down)
    EDA_norm_filtered=(EDA_second_down-EDA_min)/(EDA_max-EDA_min)
    #equation of normalisation is from
    #Hosseini, E. at al. 2022. A Low Cost EDA-based Stress Detection Using Machine Learning. 2022 IEEE International Conference on Bioinformatics and Biomedicine (BIBM). Pp. 2619-2623. Doi: 10.1109/BIBM55620.2022.9995093
    return EDA_norm_filtered

def Temp_filtering(raw_signal, sigma):
    Temp_filtered= scipy.ndimage.gaussian_filter1d(raw_signal,sigma=sigma)  
    return Temp_filtered

def Resp_filtering(raw_signal, sample_rate, method):
    resp_filtered=nk.rsp_clean(raw_signal,sample_rate, method)
    return resp_filtered

def BVP_filtering(raw_signal, sample_rate):
    filtered=bvp.bvp(raw_signal, sample_rate,show=False)
    BVP_filtered=filtered["filtered"]
    return BVP_filtered
```

```python
#Execution of filtering and segmenting the signal
#window time length refer to above

#Standard setting, 1 second overlap
s_rate=700    #standard sample rate
Window_size=s_rate*time_length  
step_size=s_rate            

#Specific window setting for downsampled EDA signal 
EDA_down_rate=14
EDA_Window_size=EDA_down_rate*time_length
EDA_step_size=EDA_down_rate

#Common processing for both chest and wrist signal
EDA_window=[]
Temp_window=[]
label_window=[]

EDA_signals=EDA_filtering(df["EDA"], s_rate)
Temp_signals=Temp_filtering(df["Temp"],4)

if(signal_source=="Chest"):
    ECG_window=[]
    EMG_window=[]
    Resp_window=[]
    ECG_signals= ECG_filtering(df["ECG"], s_rate)     
    EMG_signals=EMG_filtering(20,340,s_rate,df["EMG"])
    Resp_signals=Resp_filtering(df["Resp"],s_rate,"biosppy")
else:
    BVP_window=[]
    BVP_signals=BVP_filtering(df["BVP"],s_rate)
    
#Segmentation of signals and label
for i in range(0, len(df)-Window_size+1, step_size):
    Temp_window.append(Temp_signals[i:i+Window_size])
    label_window.append(df["label"][i:i+Window_size])
    if (signal_source=="Chest"):
        ECG_window.append(ECG_signals[i:i+Window_size])
        EMG_window.append(EMG_signals[i:i+Window_size])
        Resp_window.append(Resp_signals[i:i+Window_size])
    else:
        BVP_window.append(BVP_signals[i:i+Window_size])
    
for i in range(0, len(EDA_signals)-EDA_Window_size+1, EDA_step_size):
    if len(EDA_window)<len(label_window):
        EDA_window.append(EDA_signals[i:i+EDA_Window_size])

EDA_filtered_window=np.array(EDA_window)
Temp_filtered_window=np.array(Temp_window)
label_window=np.array(label_window)
Feature_extract={}
if(signal_source=="Chest"):
    ECG_filtered_window=np.array(ECG_window)
    EMG_filtered_window=np.array(EMG_window)
    Resp_filtered_window=np.array(Resp_window)
else:
    BVP_filtered_window=np.array(BVP_window)
```

```python
#Extract features from BVP signal for each window
if(signal_source=="Wrist"):
    Feature_extract.update({        
        "BVP_hr_mean":[],
        "BVP_hr_std":[],
        "BVP_hrv_RMSSD":[],
        "BVP_hrv_PeaktoPeak50":[],
        "BVP_hrv_pnn50":[]        })
    

    for window in BVP_filtered_window:
        peaks, _ =scipy.signal.find_peaks(window, distance=s_rate/2)
        interval=np.diff(peaks)/s_rate
        hr=60/interval
        hrv=np.diff(hr)
        hr_mean=np.mean(hr)
        hr_std=np.std(hr)
        hrv_RMSSD=np.sqrt(np.mean(np.square(hrv)))
        hrv_PeaktoPeak50=np.sum(np.abs(hrv )> 0.05)
        pnn50 = hrv_PeaktoPeak50 / len(interval)
        
        #Saving the extracted feature
        Feature_extract["BVP_hr_mean"].append(hr_mean)
        Feature_extract["BVP_hr_std"].append(hr_std)
        Feature_extract["BVP_hrv_RMSSD"].append(hrv_RMSSD)
        Feature_extract["BVP_hrv_PeaktoPeak50"].append(hrv_PeaktoPeak50)
        Feature_extract["BVP_hrv_pnn50"].append(pnn50)
```

```python
#Extract features from ECG signal for each window
if(signal_source=="Chest"):
    Feature_extract.update({        
        "ecg_hr_mean":[],
        "ecg_hr_std":[],
        "ecg_hrv_mean":[],
        "ecg_hrv_std":[],
        "ecg_hrv_rmssd":[],
        "ecg_hrv_rr50":[],
        "ecg_hrv_pnn50": [],
        "ecg_hrv_energy_LF":[],
        "ecg_hrv_energy_HF":[],
        "ecg_hrv_energy_VHF":[],
        "ecg_hrv_energy_TP":[],
        "ecg_hrv_relative_power_LF":[],
        "ecg_hrv_relative_power_HF":[],
        "ecg_hrv_relative_power_VHF":[],
        "ecg_hrv_ratio_LFHF":[],
        "ecg_hrv_norm_LF":[],
        "ecg_hrv_norm_HF":[]})
    
    for window in ECG_filtered_window:
        ecg_peak, ecg_info=nk.ecg_peaks(window, sampling_rate=s_rate)
        # heart rate and features
        rpeak=ecg_info['ECG_R_Peaks']
        rr_interval=np.diff(rpeak)/s_rate
        hr=60/rr_interval
        hr_mean=np.mean(hr)
        hr_std=np.std(hr)
        
        # Heart rate variability - Time domain
        hrv_feature_time = nk.hrv_time(ecg_peak, sampling_rate=s_rate)
        hrv_mean=hrv_feature_time["HRV_MeanNN"][0]
        hrv_std=hrv_feature_time["HRV_SDNN"][0]
        RR50=np.sum(np.abs(np.diff(rr_interval) > 0.05))
        PNN50=hrv_feature_time["HRV_pNN50"][0]
        RMSSD=hrv_feature_time["HRV_RMSSD"][0]

        #saving
        Feature_extract["ecg_hr_mean"].append(hr_mean)
        Feature_extract["ecg_hr_std"].append(hr_std)
        Feature_extract["ecg_hrv_std"].append(hrv_std)
        Feature_extract["ecg_hrv_mean"].append(hrv_mean)
        Feature_extract["ecg_hrv_rmssd"].append(RMSSD)
        Feature_extract["ecg_hrv_rr50"].append(RR50)
        Feature_extract["ecg_hrv_pnn50"].append(PNN50)
        
        #Heart rate variability - Frequency domain
        hrv_feature_freq = nk.hrv_frequency(ecg_peak, sampling_rate=s_rate, normalize=True)   
        hrv_energy_LF=hrv_feature_freq["HRV_LF"][0]
        hrv_energy_HF=hrv_feature_freq["HRV_HF"][0]
        hrv_energy_VHF=hrv_feature_freq["HRV_VHF"][0]
        hrv_energy_TP=hrv_feature_freq["HRV_TP"][0]
        hrv_relative_power_LF=hrv_energy_LF/hrv_energy_TP
        hrv_relative_power_HF=hrv_energy_HF/hrv_energy_TP
        hrv_relative_power_VHF=hrv_energy_VHF/hrv_energy_TP
        hrv_ratio_LFHF=hrv_feature_freq["HRV_LFHF"][0]
        hrv_norm_LF=hrv_feature_freq["HRV_LFn"][0]
        hrv_norm_HF=hrv_feature_freq["HRV_HFn"][0] 
        
        #Saving into dictionary
        Feature_extract["ecg_hrv_energy_LF"].append(hrv_energy_LF)
        Feature_extract["ecg_hrv_energy_HF"].append(hrv_energy_HF)
        Feature_extract["ecg_hrv_energy_VHF"].append(hrv_energy_VHF)
        Feature_extract["ecg_hrv_energy_TP"].append(hrv_energy_TP)
        Feature_extract["ecg_hrv_relative_power_LF"].append(hrv_relative_power_LF)
        Feature_extract["ecg_hrv_relative_power_HF"].append(hrv_relative_power_HF)
        Feature_extract["ecg_hrv_relative_power_VHF"].append(hrv_relative_power_VHF)
        Feature_extract["ecg_hrv_ratio_LFHF"].append(hrv_ratio_LFHF)
        Feature_extract["ecg_hrv_norm_LF"].append(hrv_norm_LF)
        Feature_extract["ecg_hrv_norm_HF"].append(hrv_norm_HF)
```

```python
#Extract features from EMG signal for each window
if(signal_source=="Chest"):
    Feature_extract.update({
        "emg_iemg":[],
        "emg_mav":[],
        "emg_mavs":[],
        "emg_ssi":[],
        "emg_var":[],
        "emg_rms":[],
        "emg_wl":[],
        "emg_energy":[],
        "emg_weight_mean_frequency":[],
        "emg_weight_median_frequency":[]})

    for window in EMG_filtered_window:

        #time doamin
        iemg=np.sum(np.abs(window))
        mav=np.mean(np.abs(window))
        mavs=np.mean(np.abs(np.diff(window)))
        ssi=np.sum(window**2)
        var=np.var(window)
        rms=np.sqrt(np.mean(window**2))
        wl=np.sum(np.abs(np.diff(window)))
        
        #saving
        Feature_extract["emg_iemg"].append(iemg)
        Feature_extract["emg_mav"].append(mav)
        Feature_extract["emg_mavs"].append(mavs)
        Feature_extract["emg_ssi"].append(ssi)
        Feature_extract["emg_var"].append(var)
        Feature_extract["emg_rms"].append(rms)
        Feature_extract["emg_wl"].append(wl)

        #frequency doamin

        N=len(window)
        fourier=scipy.fft.fft(window)    
        normalise= N/2
        fourier_frequency= scipy.fft.fftfreq(N,d=1.0/s_rate)
        index_positive= N//2
        norm_amplitude = np.abs(fourier[:index_positive])/normalise
        #Alkousa, O. 2023. Fourier Transform, the Practical Python Implementation. Available at:https://towardsdatascience.com/fourier-transform-the-practical-python-implementation-acdd32f1b96a/ [Accessed: 23 April 2025].
        
        energy=np.sum((norm_amplitude)**2)#
        weight_mean_frequency=np.average(fourier_frequency[:index_positive],weights=norm_amplitude[:index_positive])
        stack_energy=np.cumsum(norm_amplitude[:index_positive])#
        median= np.where(stack_energy>=np.sum(norm_amplitude[:index_positive])/2)[0][0]
        weight_median_frequency=fourier_frequency[:index_positive][median]

        #saving
        Feature_extract["emg_energy"].append(energy)
        Feature_extract["emg_weight_median_frequency"].append(weight_median_frequency)
        Feature_extract["emg_weight_mean_frequency"].append(weight_mean_frequency)    
```

```python
#Extract features from EDA signal for each window
Feature_extract.update( {
    "eda_mean":[],
    "eda_std":[],
    "eda_min":[],
    "eda_max":[],
    "eda_range":[],
    "eda_edl_mean":[],
    "eda_edl_std":[],
    "eda_edl_slope":[],
    "eda_edr_std":[],
    "eda_edr_num":[],
    "eda_edr_mean_amp":[],
    "eda_edr_sum_amp":[]} )

for window in EDA_filtered_window:
    #EDA stats feature 
    eda_mean=np.mean(window)
    eda_std=np.std(window)
    eda_min=np.min(window)
    eda_max=np.max(window)
    eda_range=eda_max-eda_min

    eda_process=eda.biosppy_decomposition(window,EDA_down_rate)
    #EDL feature
    eda_edl=eda_process["edl"]
    eda_edl_mean=np.mean(eda_edl)
    eda_edl_std=np.std(eda_edl)
    eda_edl_slope=np.mean(np.gradient(eda_edl))

    #EDR feature
    eda_edr_std=np.std(eda_process["edr"])
    edr_peaks, _ = scipy.signal.find_peaks(eda_process["edr"], height=0)
    eda_edr_num=len(edr_peaks)
    if eda_edr_num>0:
        eda_edr_mean_amp=np.mean(eda_process["edr"][edr_peaks])
        eda_edr_sum_amp=np.sum(eda_process["edr"][edr_peaks])
    else:
        eda_edr_mean_amp=0
        eda_edr_sum_amp=0
        
    #Saving
    Feature_extract["eda_mean"].append(eda_mean)
    Feature_extract["eda_std"].append(eda_std)
    Feature_extract["eda_min"].append(eda_min)
    Feature_extract["eda_max"].append(eda_max)
    Feature_extract["eda_range"].append(eda_range)
    Feature_extract["eda_edl_mean"].append(eda_edl_mean)
    Feature_extract["eda_edl_std"].append(eda_edl_std)
    Feature_extract["eda_edl_slope"].append(eda_edl_slope)
    Feature_extract["eda_edr_std"].append(eda_edr_std)
    Feature_extract["eda_edr_num"].append(eda_edr_num)
    Feature_extract["eda_edr_mean_amp"].append(eda_edr_mean_amp)
    Feature_extract["eda_edr_sum_amp"].append(eda_edr_sum_amp)
```

```python
#Extract features from temperature signal for each window
Feature_extract.update({
    "temp_mean":[],
    "temp_std":[],
    "temp_min":[],
    "temp_max":[],
    "temp_range":[],
    "temp_slope":[]})

for window in Temp_filtered_window:
    temp_mean=np.mean(window)
    temp_std=np.std(window)
    temp_min=np.min(window)
    temp_max=np.max(window)
    temp_range=temp_max-temp_min
    temp_slope=np.mean(np.gradient(window))
    
    #Saving
    Feature_extract["temp_mean"].append(temp_mean)
    Feature_extract["temp_std"].append(temp_std)
    Feature_extract["temp_min"].append(temp_min)
    Feature_extract["temp_max"].append(temp_max)
    Feature_extract["temp_range"].append(temp_range)
    Feature_extract["temp_slope"].append(temp_slope)
```

```python
#Extract features from respiration signal for each window
if(signal_source=="Chest"):
    Feature_extract.update({
        "resp_mean_rate":[],
        "resp_std_rate":[],
        "resp_mean_depth":[],
        "resp_P_T_ratio":[],
        "resp_mean_duration":[],
        "resp_std_duration":[]})

    for window in Resp_filtered_window:
        
        resp_peaks, resp_info=nk.rsp_peaks(window,sampling_rate=s_rate)
        resp_rate=nk.rsp_rate(window,method="troughs",sampling_rate=s_rate)
        resp_mean_rate=np.mean(resp_rate)#per mintue
        resp_std_rate=np.std(resp_rate)
        resp_amp=nk.rsp_amplitude(window, resp_peaks)
        resp_mean_depth=np.mean(resp_amp)
        
        resp_peak=0
        resp_trough=0
        for i in resp_info["RSP_Peaks"]:
            resp_peak+=resp_amp[i]
        for i in resp_info["RSP_Troughs"]:
            resp_trough+=resp_amp[i]
        resp_P_T_ratio=resp_peak/resp_trough    

        resp_duration=np.diff(resp_info["RSP_Peaks"])/s_rate
        resp_mean_duration=np.mean(resp_duration)
        resp_std_duration=np.std(resp_duration)

        #Saving
        Feature_extract["resp_mean_rate"].append(resp_mean_rate)
        Feature_extract["resp_std_rate"].append(resp_std_rate)
        Feature_extract["resp_mean_depth"].append(resp_mean_depth)
        Feature_extract["resp_P_T_ratio"].append(resp_P_T_ratio)
        Feature_extract["resp_mean_duration"].append(resp_mean_duration)
        Feature_extract["resp_std_duration"].append(resp_std_duration)
```

```python
#Finding the majority class for each window

label_class=[]

for window in label_window:
   counter=[0,0,0,0,0]
   labels,counts=np.unique(window, return_counts=True)
   for key, count in zip(labels,counts):#0=other, 1 = baseline, 2 = stress, 3 = amusement, 4= meditation
      counter[key]=count

   max_index=np.argmax(counter)
   label_class.append(max_index)

Feature_extract.update({"label":label_class})
```

```python
#Conduct validation before exporting each CSV file

if(signal_source=="Wrist"):
    verification_window=[len(BVP_filtered_window),len(EDA_filtered_window),len(Temp_filtered_window),len(label_window)]
    win_len,number=np.unique(verification_window, return_counts=True)
else:
    verification_window=[len(ECG_filtered_window),len(EMG_filtered_window),len(EDA_filtered_window),len(Temp_filtered_window),len(Resp_filtered_window),len(label_window)]
    win_len,number=np.unique(verification_window, return_counts=True)

#Checking the sliding window are in equal size
valid=True
fail_list=[]
if (len(win_len)>=2 or number[0]!=len(verification_window)):
    valid=False

#Checking the length of features are matching
for feature_label,features_array in Feature_extract.items():
    if(win_len[0]!=len(features_array)):
        valid=False
        fail_list.append(feature_label)

#Export csv file of extracted features from subject or raise Exception 
if valid and len(fail_list)==0:
    processed_frame=pd.DataFrame(Feature_extract)
    processed_frame.to_csv(CSV_filename, index=False)
else:
    raise Exception("Output length verification fail")
```
